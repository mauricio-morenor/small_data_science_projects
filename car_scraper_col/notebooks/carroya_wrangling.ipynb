```python
class CarCleaner:
    def __init__(self, filename):
        # Initialize the CarCleaner instance with a filename
        self.filename = filename
        self.data = self.load_data()

    def load_data(self):
        # Load car data from the specified CSV file
        self.data = pd.read_csv(self.filename)

    def clean_brand(self, old_column):
        # Extract the first word from the old brand column and create a new clean_brand column
        self.data["clean_brand"] = self.data[old_column].str.split(" ").str[0]

    def clean_model(self, old_column):
        # Extract the model information from the old column using a regular expression pattern
        pattern = "\s([^\\]+)\n"
        self.data["clean_model"] = [re.search(pattern, item).group(1) if re.search(pattern, item) else None for item in self.data[old_column]]

    def extract_even_plate_last_digit(self, plate_number):
        # Extract the last digit of the plate number and determine if it's even or odd
        try:
            last_digit = int(plate_number[-1])
            return 1 if last_digit % 2 == 0 else 0
        except (ValueError, IndexError):
            return 0

    def clean_tags(self, tags_column_name, city_colname, kms_colname, year_colname, plate_colname):
        # Split the tags column into separate columns for city, kilometers, year, and plate
        self.data[[city_colname, kms_colname, year_colname, plate_colname]] = self.data[tags_column_name].str.split("\n", expand=True)
        # Apply cleaning operations to plate column and numeric columns
        self.data[plate_colname] = self.data[plate_colname].apply(self.extract_even_plate_last_digit)
        self.data[kms_colname] = self.data[kms_colname].str.replace("\.", "").str.replace(" Km", "").astype(int)
        self.data[year_colname] = self.data[year_colname].astype(int)
        return self.data

    def is_automatic(self, old_column, is_automatic_column):
        # Classify if the vehicle is automatic (2), manual (1), or unknown (0)
        def classify_automatic(value):
            if "Aut" in value:
                return 2
            elif "Mec" in value:
                return 1
            else:
                return 0
        self.data[is_automatic_column] = self.data[old_column].apply(classify_automatic)
        return self.data

    def cylinder(self, old_column):
        # Extract the engine cylinder information using a regular expression pattern
        cyl_pattern = r'\n(\d+(\.\d+)?)'
        self.data["cylinder_clean"] = [re.search(cyl_pattern, item).group(1) if re.search(cyl_pattern, item) else None for item in self.data[old_column]]

    def clean_prices(self, old_column):
        # Clean and convert price values in the specified column
        self.data["clean_plate"] = self.data[old_column].str.split("\n").str[0]  # Separate values
        self.data["clean_plate"] = self.data["clean_plate"].str.replace("$", "").str.replace(".", "").astype(int)  # Remove symbols and convert to int
```

```python
#Testing the Class with the first file:
car_cleaner1 = CarCleaner('../data/carroya_usados_110823.csv')

# Apply cleaning methods
car_cleaner1.clean_brand('Tarjeta')
car_cleaner1.clean_model('Marca')
car_cleaner1.clean_tags('tags', 'City', 'Kms', 'Year', 'Plate')
car_cleaner1.is_automatic('tags', 'IsAutomatic')  # Replace with actual column names
car_cleaner1.cylinder('Tarjeta')
car_cleaner1.clean_prices('Precios')

# Print the first few rows of the cleaned DataFrame
print(car_cleaner1.data.head(3))
```
